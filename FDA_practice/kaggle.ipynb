# [Data Link](https://www.kaggle.com/ronitf/heart-disease-uci)

##### Value
| column | description |
|---|---|
| age | age in years |
| trestbps | resting blood pressure (in mm Hg on admission to the hospital) |
| chol | serum cholestoral in mg/dl |
| thalach | maximum heart rate achieved |
| oldpeak | ST depression induced by exercise relative to rest |

##### Type
| column | description |
|---|---|
| cp | chest pain type |
| restecg | resting electrocardiographic results |
| slope | the slope of the peak exercise ST segment |
| ca | number of major vessels (0-3) colored by flourosopy |
| thal | 3 = normal; 6 = fixed defect; 7 = reversable defect |

##### Boolean
| column | description |
|---|---|
| sex | (1 = male; 0 = female) |
| fbs | (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false) |
| exang | exercise induced angina (1 = yes; 0 = no) |
| target | 1 or 0 |

# Init

```python
import random
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

pd.options.display.max_rows = 10
figsize = (20,10)
```

```python
heart = pd.read_csv('heart.csv')
```

```python
cols_value = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
cols_type = ['cp', 'restecg', 'slope', 'ca', 'thal']
cols_bool = ['sex', 'fbs', 'exang', 'target']
```

# Data

```python
heart
```

# Distribution of age

```python
sns.distplot(heart['age'], kde=False);
```

# Sample size of target and non-target

```python
sns.countplot(x='target', data=heart);
```

# Relationship between cholestral and maximum heart rate

```python
sns.jointplot(x='thalach', y='chol', data=heart, kind='scatter');
```

# Impact of sex and high blood pressure on maximum heart rate

```python
g = sns.FacetGrid(heart, row='sex', col='fbs')
g.map(sns.kdeplot, 'thalach');
```

# Relationship

```python
sns.pairplot(heart[cols_value], diag_kind='kde', kind='reg');
```
